```

def Find_Resolution(dpi,dim = []):
  W = dpi*dim[0]
  H = dpi*dim[1]
  H=3*round(H/3)
  W=3*round(W/3)  
  W =round(W/3)
  H = round(H/3)
  
  return W ,H

dim = [9,13]
print(dim[0])
(W,H) = Find_Resolution(215,dim)
print(W,H)
```
